```python
import requests
import json
import pandas as pd
import csv
from pandas.io.json import json_normalize
# youtube vid for help https://youtu.be/LYh8ih2X5Oo
```

```python
url = "https://pogoapi.net/api/v1/pokemon_genders.json"
url_response = requests.get(f"{url}").json()
```

```python
output = pd.DataFrame()

for i in range(5):
    dictionary = url_response["0M_1F"][i]
    output = output.append(dictionary, ignore_index=True)
```

```python
df_3 = pd.DataFrame(url_response['0M_1F'])
pd.json_normalize(df_3['gender']).head()
```

```python
df_3sub = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in df_3['form'].items()])).T.rename(columns={0: "form"})
df_4sub = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in df_3['pokemon_id'].items()])).T.rename(columns={0: "pokemon_id"})
```

```python
df_5sub = pd.concat([df_4sub,df_3sub],axis=1)
```

```python
pd.concat([df_5sub,pd.json_normalize(df_3['gender'])],axis=1)
```
